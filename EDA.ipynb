```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset
df = pd.read_csv('COVID-19-Hospitals-Treatment-Plan.csv')
print('Shape:', df.shape)
df.head()
```

```python
df.info()
df.describe(include='all')
df.isnull().sum()
```

```python
print('Duplicate rows:', df.duplicated().sum())
df.drop_duplicates(inplace=True)

# Unique values per column
for col in df.columns:
    print(f'{col}: {df[col].nunique()} unique values')
```

```python
print("--- Descriptive Statistics ---")
print("\nSummary statistics for numerical features:")
print(df.describe())
print("\nSummary statistics for categorical features:")
print(df.describe(include='object'))
print("\n" + "="*50)
```

```python
df['Stay_Days'].value_counts().plot(kind="bar")
plt.title('Stay_Days Category Distribution')
plt.xlabel('Length of Stay Category')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# Age vs Stay_Days
plt.figure(figsize=(10,6))
sns.boxplot(x='Stay_Days', y='Age', data=df)
plt.title('Age vs Stay_Days')
plt.show()

# Severity vs Stay_Days
sns.countplot(x='Illness_Severity', hue='Stay_Days', data=df)
plt.title('Severity vs Length of Stay')
plt.show()

# Department vs Stay_Days
sns.countplot(x='Department', hue='Stay_Days', data=df)
plt.title('Department vs Stay_Days')
plt.xticks(rotation=45)
plt.show()
```

```python
categorical_features_to_plot = ['Hospital_region', 'Department', 'Ward_Type', 'Illness_Severity', 'Age']
```

```python
plt.figure(figsize=(18, 14))
for i, feature in enumerate(categorical_features_to_plot, 1):
    plt.subplot(3, 2, i)
    sns.countplot(
        y=df[feature],
        order=df[feature].value_counts().index,
        palette='coolwarm',   # Changed palette for a different color scheme
        edgecolor='black'     # Added edge color for clearer bar separation
    )
    plt.title(f'Distribution of {feature}', fontsize=14, fontweight='bold')
    plt.xlabel('Count', fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.6)  # Added light grid lines on x axis
plt.tight_layout()
plt.show()
```
